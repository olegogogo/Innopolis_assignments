```python
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1.1
##### Create two-modal distribution:

```python
data = np.random.normal(1,100,2000)
data2 = np.random.normal(400,100,3000)
data = np.concatenate((data,data2))
plt.hist(data, bins=150)
plt.show()
```

##### Taking 100 samples from distribution and compute mean values
##### Plot the results

```python
samples = []
for i in range(3000):
    sample = np.random.choice(data,100,replace=False).mean()
    samples.append(sample)
plt.hist(samples, bins=150)
plt.show()
```

### Task 1.2
##### Generate noisy data

```python
def fn(X):
    return X**2

n_samples = 200
X = np.sort(np.random.randint(0,100,n_samples))
noisyy = fn(X) + np.random.randn(n_samples) * 500
plt.scatter(X,noisyy,label="Samples")
plt.title('synthetic dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc="best")
plt.show()
```

##### Fit polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
poly = PolynomialFeatures(degree=4)
line = LinearRegression()
pipeline = Pipeline([('poly',poly),('line',line)])
pipeline.fit(X.reshape(-1,1),noisyy)
y_pred = pipeline.predict(X.reshape(-1,1))
```

##### Plot 3 graphs together

```python
plt.scatter(X, noisyy, label="Samples")
plt.plot(X, fn(X), label="True function", color = 'r')
plt.plot(X,y_pred, label="Polinomial function", color = 'g')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
